## Ejercicio 1 ‚Äì Transformada Inversa de Laplace

En este primer ejercicio aplico el m√©todo que vimos en el cuaderno para obtener la transformada inversa de Laplace de una funci√≥n racional. verificamos si la fracci√≥n es propia, hacer la descomposici√≥n en fracciones parciales y usar la tabla para obtener la se√±al en el tiempo.

La funci√≥n que quise resolver fu√©:

$$
X(s)=\frac{s+3}{(s+1)(s+2)}
$$

### 1. Fracciones parciales

Planteo la forma general:

$$
\frac{s+3}{(s+1)(s+2)}=\frac{A}{s+1}+\frac{B}{s+2}
$$

Multiplico por \((s+1)(s+2)\):

$$
s+3=A(s+2)+B(s+1)
$$

Distribuyo:

$$
s+3 = As + 2A + Bs + B
$$

Agrupo y comparo coeficientes:

Coeficiente de \(s\):
$$
A+B=1
$$

T√©rmino independiente:
$$
2A+B=3
$$

Resto ecuaciones:
$$
A=2
$$

Sustituyo en \(A+B=1\):
$$
B=-1
$$

Entonces:

$$
X(s)=\frac{2}{s+1}-\frac{1}{s+2}
$$



### 2. Transformada inversa

Uso la tabla del cuaderno:

$$
\mathcal{L}^{-1}\left\{\frac{1}{s+a}\right\}=e^{-at}u(t)
$$

Por lo tanto:

$$
\mathcal{L}^{-1}\left\{\frac{2}{s+1}\right\}=2e^{-t}u(t)
$$

$$
\mathcal{L}^{-1}\left\{\frac{1}{s+2}\right\}=e^{-2t}u(t)
$$

Como el t√©rmino tiene signo negativo:

$$
x(t)=2e^{-t}u(t)-e^{-2t}u(t)
$$

### Resultado final

$$
x(t)=\left(2e^{-t}-e^{-2t}\right)u(t)
$$


```python
import sympy as sp

s, t = sp.symbols('s t', real=True, positive=True)

Xs = (s + 3)/((s + 1)*(s + 2))
xt = sp.inverse_laplace_transform(Xs, s, t)
print("SymPy no muestra el ùë¢(ùë°) porque, por defecto, asume que todas las se√±ales obtenidas por transformada inversa son causales, y que la transformada inversa est√° definida para ùë°‚â•0.")
sp.simplify(xt)
```

## Ejercicio 2 ‚Äì Transformada Inversa de Laplace de
## $$ X(s)=\frac{1}{(s+1)(s+2)^2}$$

<br>

En este ejercicio aplico el mismo m√©todo del cuaderno para obtener la transformada inversa
de Laplace, pero ahora con la siguiente ecuacion que nos dan:

$$
X(s)=\frac{1}{(s+1)(s+2)^2}
$$

La idea es descomponer la funci√≥n en fracciones parciales, porque permite escribirla como
una suma de t√©rminos que puedo transformar directamente usando la tabla.



### 1. Fracciones parciales

Planteo la forma general:

$$
\frac{1}{(s+1)(s+2)^2} = \frac{A}{s+1} + \frac{B}{s+2} + \frac{C}{(s+2)^2}
$$

Multiplico por \((s+1)(s+2)^2\):

$$
1 = A(s+2)^2 + B(s+1)(s+2) + C(s+1)
$$

Ahora desarrollo cada parte:

- Primer t√©rmino:
$$
A(s+2)^2 = A(s^2 + 4s + 4)
$$

- Segundo t√©rmino:
$$
B(s+1)(s+2) = B(s^2 + 3s + 2)
$$

- Tercer t√©rmino:
$$
C(s+1) = Cs + C
$$

Sumo todo:

$$
1 = (A + B)s^2 + (4A + 3B + C)s + (4A + 2B + C)
$$

Comparo coeficientes con el lado izquierdo, que es:

$$
1 = 0s^2 + 0s + 1
$$

Entonces el sistema es:

1)  
$$
A + B = 0
$$

2)  
$$
4A + 3B + C = 0
$$

3)  
$$
4A + 2B + C = 1
$$

Resto las ecuaciones (3) ‚Äì (2):

$$
(4A + 2B + C) - (4A + 3B + C) = 1 - 0
$$

$$
-B = 1 \quad \Rightarrow \quad B = -1
$$

De la ecuaci√≥n 1:

$$
A + (-1) = 0 \Rightarrow A = 1
$$

Sustituyo A y B en la ecuaci√≥n 2:

$$
4(1) + 3(-1) + C = 0
$$

$$
4 - 3 + C = 0
$$

$$
C = -1
$$

Con esto, la descomposici√≥n queda:

$$
X(s) = \frac{1}{s+1} - \frac{1}{s+2} - \frac{1}{(s+2)^2}
$$

<br>

### 2. Transformada inversa

Uso la tabla:

-  
$$
\mathcal{L}^{-1}\left\{\frac{1}{s+a}\right\} = e^{-at}u(t)
$$

-  
$$
\mathcal{L}^{-1}\left\{\frac{1}{(s+a)^2}\right\} = t e^{-at}u(t)
$$

Aplico cada t√©rmino:

1)  
$$
\mathcal{L}^{-1}\left\{\frac{1}{s+1}\right\} = e^{-t}u(t)
$$

2)  
$$
\mathcal{L}^{-1}\left\{\frac{1}{s+2}\right\} = e^{-2t}u(t)
$$

3)  
$$
\mathcal{L}^{-1}\left\{\frac{1}{(s+2)^2}\right\} = t e^{-2t}u(t)
$$

El resultado final es:

$$
x(t)= e^{-t}u(t) - e^{-2t}u(t) - t e^{-2t}u(t)
$$

o agrupado:

$$
x(t)= \left( e^{-t} - e^{-2t} - t e^{-2t} \right)u(t)
$$


```python
import sympy as sp

s, t = sp.symbols('s t', real=True, positive=True)

Xs = 1 / ((s+1)*(s+2)**2)

xt = sp.inverse_laplace_transform(Xs, s, t)
print("SymPy no muestra el ùë¢(ùë°) porque, por defecto, asume que todas las se√±ales obtenidas por transformada inversa son causales, y que la transformada inversa est√° definida para ùë°‚â•0.")
sp.simplify(xt)
```

**Ejercicio 1 ‚Äì Relaci√≥n entre Œª y la salida obtenida al resolver la EDO**

<br>

En el cuaderno se trabaja con la siguiente ecuaci√≥n diferencial lineal con coeficientes
constantes:

$$
a_0 y(t) + a_1 y'(t) + a_2 y''(t) = x(t)
$$

El an√°lisis se hace usando la entrada exponencial compleja:

$$
x(t) = e^{st}
$$

Cuando esta funci√≥n se reemplaza en la EDO, la soluci√≥n obtenida en el cuaderno es:

$$
y(t) = \frac{e^{st}}{a_0 + a_1 s + a_2 s^2}
$$

Esta expresi√≥n deja ver directamente la relaci√≥n con Œª, porque esa salida puede reescribirse como:

$$
y(t) = \lambda \, e^{st}
$$

donde:

$$
\lambda = \frac{1}{a_0 + a_1 s + a_2 s^2}
$$




- La funci√≥n $(e^{st}$) vuelve a aparecer en la salida sin modificaci√≥n en su forma.
- Lo √∫nico que cambia es que ahora est√° multiplicada por un n√∫mero (Œª).
- Ese Œª depende √∫nicamente de los coeficientes del sistema y del valor de s.
- Por eso Œª es el **autovalor** asociado a la autofunci√≥n \(e^{st}\).

Œª es el factor que escala a la entrada exponencial cuando pasa por el
sistema descrito por la EDO. La salida es simplemente esa misma exponencial multiplicada
por Œª.



**Ejercicio 2 ‚Äì Generalizaci√≥n a una EDO de orden arbitrario**

En el ejercicio anterior trabaj√© con una EDO lineal de segundo orden con coeficientes constantes:

$$
a_0\,y(t) + a_1\,y'(t) + a_2\,y''(t) = e^{st}
$$

y la soluci√≥n que se obtuvo fue de la forma:

$$
y(t) = \lambda\, e^{st}, \quad \text{con} \quad
\lambda = \frac{1}{a_0 + a_1 s + a_2 s^2}
$$

Ahora la pregunta es si este mismo comportamiento se puede extender a **una EDO de orden arbitrario**, es decir, de orden \(n\).

<br>
<br>

###1. Planteo de la EDO general

Considero ahora una EDO lineal de orden \(n\), con coeficientes constantes:

$$
a_0\,y(t) + a_1\,y'(t) + a_2\,y''(t) + \dots + a_n\,y^{(n)}(t) = x(t)
$$

donde \(a_0, a_1, \dots, a_n\) son constantes reales y la entrada del sistema es nuevamente una exponencial compleja:

$$
x(t) = e^{st}
$$

La idea es verificar si, al igual que en el caso de segundo orden, la salida vuelve a ser de la forma:

$$
y(t) = \lambda\, e^{st}
$$

para alg√∫n valor de \(\lambda\) que dependa de los coeficientes del sistema y del par√°metro \(s\).

<br>
<br>

### 2. Derivadas de la exponencial

Lo primero que uso es que la derivada de una exponencial compleja es muy sencilla. Para \(x(t)=e^{st}\), se cumple:

- Primera derivada:
$$
\frac{d}{dt} e^{st} = s e^{st}
$$

- Segunda derivada:
$$
\frac{d^2}{dt^2} e^{st} = s^2 e^{st}
$$

- En general, la derivada de orden \(n\):
$$
\frac{d^n}{dt^n} e^{st} = s^n e^{st}
$$

por muchas veces que derive $(e^{st}$), la forma de la funci√≥n no cambia: siempre sigue siendo $(e^{st}$), solo que multiplicada por una potencia de $(s$).

<br>
<br>

### 3. Sustituci√≥n en la EDO

Ahora supongo que la salida tiene la forma:

$$
y(t) = \lambda\, e^{st}
$$

y sustituyo esta expresi√≥n en la EDO general. Entonces, sus derivadas ser√°n:

-  
$$
y'(t) = \lambda\, s\, e^{st}
$$

-  
$$
y''(t) = \lambda\, s^2\, e^{st}
$$

-  
$$
y^{(n)}(t) = \lambda\, s^n\, e^{st}
$$

Sustituyendo todo en la ecuaci√≥n:

$$
a_0\,y(t) + a_1\,y'(t) + a_2\,y''(t) + \dots + a_n\,y^{(n)}(t) = e^{st}
$$

queda:

$$
a_0 (\lambda e^{st}) + a_1 (\lambda s e^{st}) + a_2 (\lambda s^2 e^{st}) + \dots + a_n (\lambda s^n e^{st}) = e^{st}
$$

Factorizo \(\lambda e^{st}\):

$$
\lambda e^{st} \left(a_0 + a_1 s + a_2 s^2 + \dots + a_n s^n\right) = e^{st}
$$

Como \(e^{st} \neq 0\), puedo simplificar ambos lados por \(e^{st}\):

$$
\lambda \left(a_0 + a_1 s + a_2 s^2 + \dots + a_n s^n\right) = 1
$$

De aqu√≠ despejo \(\lambda\):

$$
\lambda = \frac{1}{a_0 + a_1 s + a_2 s^2 + \dots + a_n s^n}
$$

<br>
<br>

### 4. Interpretaci√≥n del resultado

La salida final del sistema queda:

$$
y(t) = \lambda e^{st} = \frac{e^{st}}{a_0 + a_1 s + a_2 s^2 + \dots + a_n s^n}
$$

Con esto se ve que:

- La se√±al de entrada $(e^{st}$) **vuelve a aparecer en la salida con la misma forma**.
- Lo √∫nico que cambia es un factor escalar $(\lambda$).
- Ese $(\lambda$) depende de los coeficientes de la EDO y del valor de $(s$), y es:

$$
\lambda = \frac{1}{a_0 + a_1 s + a_2 s^2 + \dots + a_n s^n}
$$


<br>

sin importar el orden de la EDO, si la entrada es \(e^{st}\), la salida conserva la misma forma y solo se escala por \(\lambda\).

## **2.12 ‚Äì Funci√≥n de transferencia en lazo abierto para circuitos RLC**

En este punto quiero encontrar la funci√≥n de transferencia en lazo abierto para:

1. Un circuito **RLC serie** (entrada: tensi√≥n de alimentaci√≥n, salida: tensi√≥n en el capacitor).
2. Un circuito **RLC en paralelo** (entrada: corriente de alimentaci√≥n, salida: corriente en el inductor).

Seg√∫n el material de apoyo, asumo que el sistema parte sin energ√≠a almacenada:
- $(v_C(0^+) = 0$)
- $(i_L(0^+) = 0$)

Es decir, las condiciones iniciales son nulas y puedo trabajar con las ecuaciones en Laplace de forma directa.

<br>

### 1) Circuito RLC serie (entrada: tensi√≥n, salida: tensi√≥n en el capacitor)

En el circuito RLC serie la misma corriente $(i(t)$) circula por los tres elementos $(R$), $(L$) y $(C$).  
La entrada es la tensi√≥n de alimentaci√≥n $(x(t) = v_{\text{in}}(t)$) y la salida es la tensi√≥n en el capacitor $(y(t) = v_C(t)$).

La relaci√≥n en el tiempo es:

$$
v_{\text{in}}(t) = v_R(t) + v_L(t) + v_C(t)
$$

Uso las relaciones de los elementos:

- Resistor: $(v_R(t) = R\,i(t)$)
- Inductor: $(v_L(t) = L\,\dfrac{di(t)}{dt}$)
- Capacitor: $(i(t) = C\,\dfrac{dv_C(t)}{dt}$)

Reemplazo $(i(t)$) en funci√≥n de $(v_C(t)$):

- $(i(t) = C\,\dfrac{dv_C}{dt}$)
- $(\dfrac{di}{dt} = C\,\dfrac{d^2 v_C}{dt^2}$)

Entonces:

- $(v_R(t) = R C \dfrac{dv_C}{dt}$)
- $(v_L(t) = L C \dfrac{d^2 v_C}{dt^2}$)
- $(v_C(t) = v_C(t)$)

La ecuaci√≥n queda:

$$
v_{\text{in}}(t) = L C \frac{d^2 v_C(t)}{dt^2} + R C \frac{dv_C(t)}{dt} + v_C(t)
$$

Identificando entrada y salida:

$$
x(t) = v_{\text{in}}(t), \quad y(t) = v_C(t)
$$

se puede escribir:

$$
L C\, y''(t) + R C\, y'(t) + y(t) = x(t)
$$

<br>
<br>

#### Funci√≥n de transferencia en Laplace

Con condiciones iniciales nulas, al aplicar la transformada de Laplace:

$$
(L C s^2 + R C s + 1)\,Y(s) = X(s)
$$

La funci√≥n de transferencia en lazo abierto es:

$$
H_{\text{serie}}(s) = \frac{Y(s)}{X(s)} = \frac{V_C(s)}{V_{\text{in}}(s)}
= \frac{1}{L C s^2 + R C s + 1}
$$

<br>

### 2) Circuito RLC en paralelo (entrada: corriente, salida: corriente en el inductor)

Ahora considero un circuito RLC en paralelo. Todos los elementos comparten el mismo voltaje \(v(t)\).

- Entrada: corriente de alimentaci√≥n $(x(t) = i_{\text{in}}(t)$).
- Salida: corriente en el inductor $(y(t) = i_L(t)$).

En el dominio de Laplace, usando impedancias/admitancias con condiciones iniciales nulas:

- Resistor: $(Z_R = R \Rightarrow Y_R = \dfrac{1}{R}$)
- Inductor: $(Z_L = sL \Rightarrow Y_L = \dfrac{1}{sL}$)
- Capacitor: $(Z_C = \dfrac{1}{sC} \Rightarrow Y_C = sC$)

La corriente de entrada se reparte entre las tres ramas:

$$
I_{\text{in}}(s) = I_R(s) + I_L(s) + I_C(s)
$$

Cada corriente se escribe en funci√≥n del voltaje de nodo $(V(s)$):

- $(I_R(s) = Y_R V(s) = \dfrac{1}{R} V(s)$)
- $(I_L(s) = Y_L V(s) = \dfrac{1}{sL} V(s)$)
- $(I_C(s) = Y_C V(s) = s C\, V(s)$)

Entonces:

$$
I_{\text{in}}(s) = \left( \frac{1}{R} + \frac{1}{sL} + s C \right) V(s)
$$

La salida es la corriente en el inductor:

$$
Y(s) = I_L(s) = \frac{1}{sL} V(s)
$$

<br>

#### Funci√≥n de transferencia en Laplace

Escribo la funci√≥n de transferencia como:

$$
H_{\text{paralelo}}(s) = \frac{Y(s)}{X(s)} = \frac{I_L(s)}{I_{\text{in}}(s)}
= \frac{\dfrac{1}{sL} V(s)}{\left( \dfrac{1}{R} + \dfrac{1}{sL} + s C \right) V(s)}
$$

Simplificando:

$$
H_{\text{paralelo}}(s)
= \frac{1/(sL)}{1/R + 1/(sL) + sC}
$$

Al llevar todo a un denominador com√∫n y simplificar, se obtiene:

$$
H_{\text{paralelo}}(s) = \frac{R}{C L R s^2 + L s + R}
$$

Dividiendo numerador y denominador por \(R\):

$$
H_{\text{paralelo}}(s)
= \frac{1}{L C s^2 + \frac{L}{R} s + 1}
$$

## **2.13 ‚Äì Salida en el tiempo para un circuito RLC en lazo cerrado**

**Parto de la funci√≥n de transferencia en lazo abierto que obtuve en el punto anterior para el
circuito RLC serie:**

$$
H(s) = \frac{1}{L C s^2 + R C s + 1}
$$

El ejercicio nos pide encontrar la respuesta del sistema pero **en configuraci√≥n de lazo cerrado**.
Como se muestra en el cuaderno de Fracciones Parciales del RLC, la funci√≥n de
transferencia en lazo cerrado queda:

$$
H_{lc}(s) = \frac{H(s)}{1 + H(s)}
$$

Reemplazo \(H(s)\):

$$
H_{lc}(s) = \frac{\frac{1}{LCs^2 + RCs + 1}}
{1 + \frac{1}{LCs^2 + RCs + 1}}
$$

Llevo todo al mismo denominador:

$$
H_{lc}(s) =
\frac{1}{LCs^2 + RCs + 2}
$$

De esta forma, la din√°mica del sistema de lazo cerrado tiene un denominador ligeramente
modificado (donde aparece el ‚Äú+2‚Äù), que ya est√° explicado en el cuaderno.

A partir de esta funci√≥n de transferencia encontrar√© la salida para tres entradas:

1. impulso  
2. escal√≥n unitario  
3. rampa  

<br>

### 1) Entrada impulso

Para la entrada impulso:

$$
X(s)=1
$$

La salida en Laplace es:

$$
Y(s) = H_{lc}(s)\, X(s) = \frac{1}{LCs^2 + RCs + 2}
$$

Esta expresi√≥n coincide con la forma general de un sistema de segundo orden.  
Dependiendo del valor de \(R\), el sistema puede ser subamortiguado, cr√≠tico o
sobreamortiguado. Para mantener la expresi√≥n general, escribo el denominador como:

$$
LCs^2 + RCs + 2 = LC\left(s^2 + \frac{R}{L}s + \frac{2}{LC}\right)
$$

La respuesta al impulso es la transformada inversa de esa expresi√≥n.  
En forma general, queda:

$$
y_{\delta}(t) = \text{Respuesta natural seg√∫n los polos del sistema}
$$



<br>

### 2) Entrada escal√≥n unitario

Para el escal√≥n unitario:

$$
X(s)=\frac{1}{s}
$$

La salida es:

$$
Y(s)= H_{lc}(s)\, \frac{1}{s}
= \frac{1}{s(LCs^2 + RCs + 2)}
$$

Para obtener la salida en el tiempo uso fracciones parciales, igual que en el cuaderno:

$$
\frac{1}{s(LCs^2 + RCs + 2)}
= \frac{A}{s} + \frac{Bs + C}{LCs^2 + RCs + 2}
$$

De aqu√≠ se obtiene la t√≠pica expresi√≥n:

$$
y_{u}(t)= A + e^{-\alpha t}\left(B\cos(\omega t) + C\sin(\omega t)\right)
$$


<br>

### 3) Entrada rampa

La rampa en Laplace es:

$$
X(s)=\frac{1}{s^2}
$$

Entonces:

$$
Y(s)= H_{lc}(s)\,\frac{1}{s^2}
= \frac{1}{s^2 (LCs^2 + RCs + 2)}
$$

Igual que en el cuaderno, hago fracciones parciales:

$$
\frac{1}{s^2(LCs^2 + RCs + 2)}
= \frac{A}{s} + \frac{B}{s^2} + \frac{Cs + D}{LCs^2 + RCs + 2}
$$

La transformada inversa da la forma general:

- un t√©rmino proporcional al tiempo (por la rampa),
- un t√©rmino constante,
- m√°s t√©rminos exponenciales y sinusoidales amortiguados.

La se√±al queda:

$$
y_r(t) = Bt + A + e^{-\alpha t}\left(C\cos(\omega t) + D\sin(\omega t)\right)
$$






```python
import sympy as sp

# Defino variables simb√≥licas
t, s = sp.symbols('t s', real=True)
R, L, C = sp.symbols('R L C', positive=True)

# Funci√≥n de transferencia en lazo abierto (RLC serie, salida tensi√≥n en C)
H_abierto = 1 / (L*C*s**2 + R*C*s + 1)

# Funci√≥n de transferencia en lazo cerrado con realimentaci√≥n unitaria
H_cerrado = sp.simplify(H_abierto / (1 + H_abierto))
H_cerrado
```

```python
# Entradas en Laplace
X_impulso = 1        # Œ¥(t)  ->  1
X_escalon = 1/s      # u(t)  ->  1/s
X_rampa   = 1/s**2   # rampa ->  1/s^2
```

```python
# Salidas en Laplace
Y_impulso = sp.simplify(H_cerrado * X_impulso)
Y_escalon = sp.simplify(H_cerrado * X_escalon)
Y_rampa   = sp.simplify(H_cerrado * X_rampa)

Y_impulso, Y_escalon, Y_rampa
```

```python
# Transformadas inversas: salidas en el tiempo
y_impulso = sp.simplify(sp.inverse_laplace_transform(Y_impulso, s, t))
y_escalon = sp.simplify(sp.inverse_laplace_transform(Y_escalon, s, t))
y_rampa   = sp.simplify(sp.inverse_laplace_transform(Y_rampa, s, t))

y_impulso, y_escalon, y_rampa
```

en el inductor) obtuve la siguiente funci√≥n de transferencia en lazo abierto:**

$[
H(s) = \frac{1}{LCs^2 + \frac{L}{R}s + 1}
$]

Ahora el ejercicio 2.13 pide trabajar el sistema en **lazo cerrado** y encontrar la salida
en el tiempo para tres tipos de entrada: impulso, escal√≥n y rampa.

### 0) Funci√≥n de transferencia en lazo cerrado

Uso realimentaci√≥n unitaria:

$[
H_{lc}(s) = \frac{H(s)}{1 + H(s)}
$]

Reemplazo $(H(s)$):

$[
H_{lc}(s) =
\frac{\dfrac{1}{LCs^2 + \frac{L}{R}s + 1}}
     {1 + \dfrac{1}{LCs^2 + \frac{L}{R}s + 1}}
$]

Llevo todo al mismo denominador:

$[
H_{lc}(s) =
\frac{1}{LCs^2 + \frac{L}{R}s + 2}
$]

Para simplificar un poco la notaci√≥n, escribo el denominador como un polinomio gen√©rico:

$[
a_2 s^2 + a_1 s + a_0
$]

donde:

$[
a_2 = LC, \quad a_1 = \frac{L}{R}, \quad a_0 = 2
$]

Entonces:

$[
H_{lc}(s) = \frac{1}{a_2 s^2 + a_1 s + a_0}
$]

<br>

## 1) Entrada impulso

Para un **impulso**, la transformada de Laplace es:

$[
X_\delta(s) = 1
$]

La salida en Laplace es simplemente:

$[
Y_\delta(s) = H_{lc}(s)\,X_\delta(s)
             = \frac{1}{a_2 s^2 + a_1 s + a_0}
$]

Aqu√≠ no necesito fracciones parciales porque solo tengo un polinomio de segundo orden
en el denominador. Esta es la t√≠pica funci√≥n de un sistema de segundo orden. Dependiendo
de los valores de \(R, L, C\), los polos pueden ser reales o complejos.

En forma general, la respuesta al impulso se puede escribir como:

- **Sobreamortiguado (polos reales distintos):**
  $[
  y_\delta(t) = K_1 e^{p_1 t} + K_2 e^{p_2 t}
  $]

- **Subamortiguado (polos complejos conjugados):**
  $[
  y_\delta(t) = e^{-\alpha t}\left(A \cos(\omega t) + B \sin(\omega t)\right)
  $]

donde $(p_1, p_2$) o bien $(\alpha, \omega$) salen de los polos del denominador
$(a_2 s^2 + a_1 s + a_0$).

<br>



## 2) Entrada escal√≥n unitario ‚Äì Fracciones parciales paso a paso

Para el **escal√≥n unitario**, la transformada es:

$[
X_u(s) = \frac{1}{s}
$]

Entonces la salida en Laplace es:

$[
Y_u(s) = H_{lc}(s)\,X_u(s)
       = \frac{1}{s(a_2 s^2 + a_1 s + a_0)}
$]

Aqu√≠ s√≠ aplico fracciones parciales. Propongo la descomposici√≥n:

$[
\frac{1}{s(a_2 s^2 + a_1 s + a_0)}
= \frac{A}{s} + \frac{B s + C}{a_2 s^2 + a_1 s + a_0}
$]

Ahora multiplico ambos lados por \(s(a_2 s^2 + a_1 s + a_0)\):

$[
1 = A(a_2 s^2 + a_1 s + a_0) + (B s + C)s
$]

Desarrollo:

$[
1 = A a_2 s^2 + A a_1 s + A a_0 + B s^2 + C s
$]

Agrupo por potencias de \(s\):

- T√©rmino en \(s^2\):
  $[
  (A a_2 + B)s^2
  $]

- T√©rmino en \(s\):
  $[
  (A a_1 + C)s
  $]

- T√©rmino constante:
  $[
  A a_0
  $]

Como en el lado izquierdo solo tengo un **1**, eso equivale a:

$[
1 = 0 s^2 + 0 s + 1
$]

Entonces igualo coeficientes:

$[
\begin{cases}
A a_2 + B = 0 \\
A a_1 + C = 0 \\
A a_0 = 1
\end{cases}
$]

De la √∫ltima ecuaci√≥n obtengo:

$[
A = \frac{1}{a_0}
$]

Luego:

$[
B = -A a_2 = -\frac{a_2}{a_0}, \qquad
C = -A a_1 = -\frac{a_1}{a_0}
$]

Por lo tanto, la descomposici√≥n queda:

$[
\frac{1}{s(a_2 s^2 + a_1 s + a_0)}
= \frac{1}{a_0}\,\frac{1}{s}
  - \frac{a_2}{a_0}\,\frac{s}{a_2 s^2 + a_1 s + a_0}
  - \frac{a_1}{a_0}\,\frac{1}{a_2 s^2 + a_1 s + a_0}
$]

Es decir:

$[
Y_u(s)
= \frac{1}{a_0}\,\frac{1}{s}
  - \frac{a_2}{a_0}\,\frac{s}{a_2 s^2 + a_1 s + a_0}
  - \frac{a_1}{a_0}\,\frac{1}{a_2 s^2 + a_1 s + a_0}
$]

<br>

### Transformada inversa (idea general)

- El primer t√©rmino:
  $[
  \frac{1}{a_0}\,\frac{1}{s} \quad \Rightarrow \quad \frac{1}{a_0} u(t)
  $]

- Los otros dos t√©rminos tienen el mismo denominador $(a_2 s^2 + a_1 s + a_0$).
  Completando el cuadrado y usando la tabla del cuaderno, se convierten en una
  combinaci√≥n de exponenciales y senos/cosenos amortiguados.

En forma general la salida para el escal√≥n se puede escribir como:

$[
y_u(t) = \frac{1}{a_0} + e^{-\alpha t}\left(B_1 \cos(\omega t) + B_2 \sin(\omega t)\right)
$]

donde $(\alpha$) y $(\omega$) dependen de $(a_0, a_1, a_2$) y por lo tanto de $(R, L, C$).

Si ahora reemplazo $(a_0 = 2$), $(a_1 = \frac{L}{R}$), $(a_2 = LC$), la expresi√≥n queda
espec√≠ficamente en funci√≥n de \(R, L, C\) como pide el ejercicio.

<br>

## 3) Entrada rampa ‚Äì Fracciones parciales m√°s detalladas

Para la **rampa**, la transformada es:

$[
X_r(s) = \frac{1}{s^2}
$]

Entonces:

$[
Y_r(s) = H_{lc}(s)\,X_r(s)
       = \frac{1}{s^2(a_2 s^2 + a_1 s + a_0)}
$]

Planteo ahora una descomposici√≥n m√°s completa:

$[
\frac{1}{s^2(a_2 s^2 + a_1 s + a_0)}
= \frac{A}{s} + \frac{B}{s^2} + \frac{C s + D}{a_2 s^2 + a_1 s + a_0}
$]

Multiplico ambos lados por \(s^2(a_2 s^2 + a_1 s + a_0)\):

$[
1 = A s (a_2 s^2 + a_1 s + a_0)
    + B (a_2 s^2 + a_1 s + a_0)
    + (C s + D)s^2
$]

Desarrollo cada parte:

- $(A s (a_2 s^2 + a_1 s + a_0) = A(a_2 s^3 + a_1 s^2 + a_0 s)$)
- $(B (a_2 s^2 + a_1 s + a_0) = B a_2 s^2 + B a_1 s + B a_0$)
- $((C s + D)s^2 = C s^3 + D s^2$)

Sumo todo y agrupo por potencias de $(s$):

$[
1 =
(A a_2 + C)s^3
+ (A a_1 + B a_2 + D)s^2
+ (A a_0 + B a_1)s
+ (B a_0)
$]

Esto debe ser igual al polinomio constante 1:

$[
1 = 0 s^3 + 0 s^2 + 0 s + 1
$]

Entonces obtengo el sistema:

$[
\begin{cases}
A a_2 + C = 0 \\
A a_1 + B a_2 + D = 0 \\
A a_0 + B a_1 = 0 \\
B a_0 = 1
\end{cases}
$]

De la √∫ltima ecuaci√≥n:

$[
B = \frac{1}{a_0}
$]

Con esto, uso la tercera ecuaci√≥n:

$[
A a_0 + \frac{1}{a_0} a_1 = 0
\quad \Rightarrow \quad
A = -\frac{a_1}{a_0^2}
$]

Ahora la primera ecuaci√≥n:

$[
A a_2 + C = 0
\quad \Rightarrow \quad
C = -A a_2 = \frac{a_1 a_2}{a_0^2}
$]

Finalmente, la segunda ecuaci√≥n:

$[
A a_1 + B a_2 + D = 0
$]

Sustituyo \(A\) y \(B\):

$[
-\frac{a_1}{a_0^2} a_1 + \frac{1}{a_0} a_2 + D = 0
$]

$[
-\frac{a_1^2}{a_0^2} + \frac{a_2}{a_0} + D = 0
$]

$[
D = \frac{a_1^2}{a_0^2} - \frac{a_2}{a_0}
    = \frac{a_1^2 - a_0 a_2}{a_0^2}
$]

Por lo tanto, la descomposici√≥n final es:

$[
\frac{1}{s^2(a_2 s^2 + a_1 s + a_0)}
=
\frac{A}{s} + \frac{B}{s^2} + \frac{C s + D}{a_2 s^2 + a_1 s + a_0}
$]

con:

$[
A = -\frac{a_1}{a_0^2}, \quad
B = \frac{1}{a_0}, \quad
C = \frac{a_1 a_2}{a_0^2}, \quad
D = \frac{a_1^2 - a_0 a_2}{a_0^2}
$]

<br>

### Transformada inversa para la rampa (estructura general)

La transformada inversa de cada t√©rmino ser√≠a:

- $(\dfrac{B}{s^2} \Rightarrow B\,t\,u(t)$)  ‚Üí aqu√≠ aparece el t√©rmino de rampa.
- $(\dfrac{A}{s} \Rightarrow A\,u(t)$)      ‚Üí t√©rmino constante.
- Los t√©rminos con $((C s + D)/(a_2 s^2 + a_1 s + a_0)$) se convierten, como en el caso
  anterior, en exponenciales amortiguadas (con senos y cosenos).

Por eso la respuesta a la rampa tiene la forma:

$[
y_r(t) = B t + A + e^{-\alpha t}\left(C_1 \cos(\omega t) + C_2 \sin(\omega t)\right)
$]

Al final, si reemplazo $(a_0 = 2$), $(a_1 = \dfrac{L}{R}$), $(a_2 = LC$), todos los
coeficientes quedan en funci√≥n directa de $(R, L, C$), tal como pide el ejercicio.




```python
import sympy as sp

# Variables simb√≥licas
t, s = sp.symbols('t s', real=True)
R, L, C = sp.symbols('R L C', positive=True)

# Funci√≥n de transferencia en lazo cerrado para RLC en paralelo:
# H_lc(s) = 1 / (L*C*s**2 + (L/R)*s + 2)
H_lc_par = 1 / (L*C*s**2 + (L/R)*s + 2)

H_lc_par
```

1) ¬øQu√© es un sistema subamortiguado, sobreamortiguado y cr√≠ticamente amortiguado?

Un sistema de segundo orden est√° definido principalmente por:

    Œ∂  -> factor de amortiguamiento
    œâ‚Çô -> frecuencia natural no amortiguada

Dependiendo de Œ∂, el comportamiento es:

- Subamortiguado (0 < Œ∂ < 1):
  El sistema s√≠ tiene amortiguamiento pero no alcanza a matar la oscilaci√≥n de una.
  La respuesta oscila un par de veces y luego se estabiliza. Es el caso t√≠pico
  de un resorte o p√©ndulo con algo de fricci√≥n: baja y sube varias veces antes
  de quedarse quieto.

- Cr√≠ticamente amortiguado (Œ∂ = 1):
  Es la frontera entre oscilar y volverse demasiado lento. El sistema ya no
  oscila, pero llega lo m√°s r√°pido posible al valor final sin sobrepaso.
  Es el ‚Äúcaso ideal‚Äù si uno quiere rapidez sin oscilaciones.

- Sobreamortiguado (Œ∂ > 1):
  Hay demasiado amortiguamiento. El sistema no oscila, pero su respuesta
  es m√°s lenta que en el caso cr√≠tico. Es como un sistema con demasiada
  fricci√≥n, que simplemente se va moviendo despacio hasta estabilizarse.

Tambi√©n se puede hablar de sistemas inestables si Œ∂ < 0, en cuyo caso la
respuesta crece en vez de morir.

<br>

2) Expresiones de Œ∂, œâ‚Çô y œâ_d en funci√≥n de R, L y C


De los cuadernos de an√°lisis de redes RLC y del p√©ndulo el√°stico se tiene:

Frecuencia natural no amortiguada (para ambos casos, serie y paralelo):

    œâ‚Çô = sqrt( 1 / (L * C) )

A partir de ah√≠, el factor de amortiguamiento depende del tipo de circuito:

- Circuito RLC SERIE:
    Œ∂ = (R / 2) * sqrt( C / L )

  Esta f√≥rmula sale al comparar el denominador de la funci√≥n de transferencia
  del RLC serie con la forma can√≥nica:
      s^2 + 2Œ∂œâ‚Çô s + œâ‚Çô^2

- Circuito RLC PARALELO:
    Œ∂ = (1 / (2 * R)) * sqrt( L / C )

  En este caso la contribuci√≥n de la resistencia es distinta porque R est√°
  en paralelo con L y C, pero al final se llega a una expresi√≥n similar,
  solo que R aparece en el denominador.

Frecuencia natural amortiguada (solo cuando 0 < Œ∂ < 1):

    œâ_d = œâ‚Çô * sqrt( 1 - Œ∂^2 )

Esta œâ_d es la frecuencia a la que realmente oscilan las se√±ales en la
respuesta temporal cuando el sistema es subamortiguado.

```python
%%writefile app.py
# En este archivo va TODO el dashboard del ejercicio 2.14.
# La idea es seguir el mismo estilo de los cuadernos:
# - imports claros
# - uso de sympy como "sym"
# - comentarios explicando en palabras normales qu√© hace cada bloque.

import streamlit as st               # para el dashboard
import numpy as np                   # para c√°lculos num√©ricos
import sympy as sym                  # para c√°lculo simb√≥lico (Laplace, polos, etc.)
import matplotlib
matplotlib.use("Agg")               # backend "sin pantalla" para que funcione en servidor
import matplotlib.pyplot as plt      # para gr√°ficas

# Inicializamos impresi√≥n bonita de sympy (como en los cuadernos)
sym.init_printing()

# Definimos las variables simb√≥licas globales que vamos a usar
t, s = sym.symbols('t s', real=True)

#
# FUNTIONES AUXILIARES (TODO EL MODELO RLC)
#

def parametros_RLC_serie(zeta, wn, C=1.0):
    """
    Calcula R y L para un circuito RLC SERIE en lazo cerrado.
    La relaci√≥n viene del modelo de segundo orden:

        Denominador normalizado: s^2 + 2*zeta*wn*s + wn^2

    Para el RLC serie (seg√∫n los cuadernos de an√°lisis de redes):

        H(s) = 1 / (L*C*s^2 + R*C*s + 2)

    Igualando coeficientes:
        L*C = 2 / wn^2
        R*C = 2*zeta*wn*L*C

    De ah√≠ despejamos L y R.
    """

    # Primero calculamos L a partir de LC = 2/wn^2
    L = 2.0 / (wn**2 * C)

    # Luego R usando zeta = R / (2*L*wn)
    R = 2.0 * zeta * L * wn

    return float(R), float(L), float(C)


def parametros_RLC_paralelo(zeta, wn, C=1.0):
    """
    Calcula R y L para un circuito RLC PARALELO en lazo cerrado.

    Para el paralelo (seg√∫n los cuadernos de RLC paralelo):

        H(s) = 1 / (L*C*s^2 + (L/R)*s + 2)

    En forma can√≥nica de segundo orden:
        L*C = 2 / wn^2
        (L/R) = 2*zeta*wn*L*C

    Y tambi√©n se usa la relaci√≥n est√°ndar:
        zeta = 1 / (2*R*C*wn)

    Con eso despejamos R y L.
    """

    # Igual que en serie, L viene de LC = 2/wn^2
    L = 2.0 / (wn**2 * C)

    # Evitamos divisi√≥n por cero en el caso zeta = 0
    if zeta == 0:
        zeta = 1e-6

    # De zeta = 1/(2*R*C*wn) despejamos R
    R = 1.0 / (2.0 * zeta * C * wn)

    return float(R), float(L), float(C)


def H_lazo_cerrado(tipo_circuito, R, L, C):
    """
    Construye la funci√≥n de transferencia en lazo cerrado H(s)
    usando las expresiones del punto 2.12 de los cuadernos.

    - Para SERIE:
        H(s) = 1 / (L*C*s^2 + R*C*s + 2)

    - Para PARALELO:
        H(s) = 1 / (L*C*s^2 + (L/R)*s + 2)
    """

    if tipo_circuito == "Serie":
        Hs = 1 / (L*C*s**2 + R*C*s + 2)
    else:  # Paralelo
        Hs = 1 / (L*C*s**2 + (L/R)*s + 2)

    # Simplificamos por si se puede
    return sym.simplify(Hs)


def respuestas_tiempo(Hs):
    """
    A partir de H(s) calculamos las respuestas en el tiempo
    para tres entradas est√°ndar:

    - Impulso:   X(s) = 1
    - Escal√≥n:   X(s) = 1/s
    - Rampa:     X(s) = 1/s^2

    Igual que en los cuadernos de Respuesta al Impulso
    y Respuesta al Escal√≥n, usamos la transformada inversa
    de Laplace de sympy.
    """

    # Entradas en el dominio s
    X_impulso = 1
    X_escalon = 1/s
    X_rampa   = 1/s**2

    # Salidas en s
    Y_impulso = sym.simplify(Hs * X_impulso)
    Y_escalon = sym.simplify(Hs * X_escalon)
    Y_rampa   = sym.simplify(Hs * X_rampa)

    # Transformada inversa de Laplace para cada caso
    y_impulso_t = sym.simplify(sym.inverse_laplace_transform(Y_impulso, s, t))
    y_escalon_t = sym.simplify(sym.inverse_laplace_transform(Y_escalon, s, t))
    y_rampa_t   = sym.simplify(sym.inverse_laplace_transform(Y_rampa, s, t))

    return y_impulso_t, y_escalon_t, y_rampa_t


def datos_bode(Hs, w_min=0.1, w_max=100, puntos=400):
    """
    Genera los datos para el diagrama de Bode (magnitud y fase),
    igual que en el cuaderno de fracciones parciales RLC.

    Evaluamos H(jw) para un rango logar√≠tmico de frecuencias.
    """

    # Frecuencia logar√≠tmica
    w_vec = np.logspace(np.log10(w_min), np.log10(w_max), puntos)

    # s = j*w
    s_eval = 1j * w_vec

    # Convertimos H(s) simb√≥lico a funci√≥n num√©rica
    H_func = sym.lambdify(s, Hs, "numpy")

    # Evaluamos
    H_eval = H_func(s_eval)

    # Magnitud en dB y fase en grados
    magnitud = 20 * np.log10(np.abs(H_eval))
    fase     = np.angle(H_eval, deg=True)

    return w_vec, magnitud, fase


def polos_y_ceros(Hs):
    """
    Calcula polos y ceros de H(s) usando sympy.
    Esto es lo mismo que hacemos en los cuadernos
    cuando factorizamos el denominador.

    Regresa listas de n√∫meros complejos para ceros y polos.
    """

    num, den = sym.fraction(Hs)     # numerador y denominador simb√≥licos
    num_poly = sym.Poly(num, s)
    den_poly = sym.Poly(den, s)

    # Si el numerador es de grado 0, no hay ceros
    if num_poly.degree() > 0:
        ceros = [complex(c) for c in num_poly.nroots()]
    else:
        ceros = []

    polos = [complex(p) for p in den_poly.nroots()]

    return ceros, polos


def tiempos_segundo_orden(zeta, wn):
    """
    Calcula tiempos t√≠picos de un sistema de segundo orden
    como se ve en el cuaderno del p√©ndulo el√°stico / SLITs:

      - tr: tiempo de levantamiento aproximado
      - Mp: m√°ximo sobreimpulso
      - tp: tiempo del m√°ximo sobreimpulso
      - ts: tiempo de establecimiento (aprox. 2%)
      - wd: frecuencia natural amortiguada (si aplica)
    """

    # Tiempo de levantamiento aproximado
    tr = 1.8 / wn

    if 0 < zeta < 1:
        # Frecuencia amortiguada
        wd = wn * np.sqrt(1 - zeta**2)

        # M√°ximo sobreimpulso (en tanto unitario)
        Mp = np.exp(-zeta * np.pi / np.sqrt(1 - zeta**2))

        # Tiempo del m√°ximo sobreimpulso
        tp = np.pi / wd
    else:
        wd = np.nan
        Mp = 0.0
        tp = np.nan

    # Tiempo de establecimiento (criterio 2%)
    if zeta > 0:
        ts = 4.0 / (zeta * wn)
    else:
        ts = np.nan

    return tr, Mp, tp, ts, wd


#
# INTERFAZ STREAMLIT (DASHBOARD)
#

# T√≠tulo principal
st.title("Ejercicio 2.14 ‚Äì Sistemas RLC serie y paralelo")

st.markdown(
    """
    Dashboard para explorar un sistema RLC en **lazo cerrado** (serie y paralelo)
    con diferentes tipos de amortiguamiento.
    Aqu√≠ se pueden ver las respuestas al impulso, escal√≥n y rampa,
    adem√°s de Bode, polos y ceros, y los tiempos t√≠picos de segundo orden.
    """
)

# ------------------ Panel lateral (par√°metros) ------------------
st.sidebar.header("Par√°metros del sistema")

# Tipo de circuito (serie o paralelo)
tipo_circ = st.sidebar.selectbox("Tipo de circuito", ["Serie", "Paralelo"])

# Tipo de amortiguamiento
tipo_amort = st.sidebar.selectbox(
    "Tipo de amortiguamiento",
    [
        "Subamortiguado (0 < Œ∂ < 1)",
        "Sobreamortiguado (Œ∂ > 1)",
        "Cr√≠tico (Œ∂ = 1)",
        "Inestable (Œ∂ < 0)"
    ]
)

# Seg√∫n el tipo, definimos el rango del slider de zeta
if "Subamortiguado" in tipo_amort:
    z_min, z_max, z_def = 0.05, 0.95, 0.4
elif "Sobreamortiguado" in tipo_amort:
    z_min, z_max, z_def = 1.05, 3.0, 1.5
elif "Cr√≠tico" in tipo_amort:
    z_min, z_max, z_def = 1.0, 1.0, 1.0
else:  # Inestable
    z_min, z_max, z_def = -0.8, -0.05, -0.3

zeta_val = st.sidebar.slider("Œ∂ (factor de amortiguamiento)", z_min, z_max, z_def)
wn_val   = st.sidebar.slider("œâ‚Çô (frecuencia natural)", 0.5, 10.0, 2.0)
C_val    = st.sidebar.number_input("Capacitancia C [F]", value=1.0)

# ------------------ C√°lculo de R, L, C ------------------
if tipo_circ == "Serie":
    R_val, L_val, C_val = parametros_RLC_serie(zeta_val, wn_val, C_val)
else:
    R_val, L_val, C_val = parametros_RLC_paralelo(zeta_val, wn_val, C_val)

st.subheader("Par√°metros f√≠sicos estimados (a partir de Œ∂ y œâ‚Çô)")

colR, colL, colC = st.columns(3)
colR.metric("R [Œ©]", f"{R_val:.4g}")
colL.metric("L [H]", f"{L_val:.4g}")
colC.metric("C [F]", f"{C_val:.4g}")

# ------------------ Funci√≥n de transferencia H(s) ------------------
Hs = H_lazo_cerrado(tipo_circ, R_val, L_val, C_val)

# Extraemos de nuevo Œ∂ y œâ‚Çô reales a partir del denominador (por si acaso)
_, den = sym.fraction(Hs)
den_poly = sym.Poly(den, s)
a2, a1, a0 = [float(c) for c in den_poly.all_coeffs()]   # a2*s^2 + a1*s + a0

wn_real   = np.sqrt(a0 / a2)
zeta_real = a1 / (2.0 * wn_real * a2)
tr, Mp, tp, ts, wd = tiempos_segundo_orden(zeta_real, wn_real)

st.subheader("Par√°metros de segundo orden obtenidos de H(s)")

colWn, colZ, colWd = st.columns(3)
colWn.metric("œâ‚Çô (real)", f"{wn_real:.4g} rad/s")
colZ.metric("Œ∂ (real)", f"{zeta_real:.4g}")

if 0 < zeta_real < 1:
    colWd.metric("œâ_d", f"{wd:.4g} rad/s")
else:
    colWd.metric("œâ_d", "N/A")

# ------------------ Respuestas en el tiempo ------------------
st.subheader("Respuestas en el tiempo para distintas entradas")

with st.spinner("Calculando Laplace inversa (puede tardar unos segundos)..."):
    y_imp_t, y_esc_t, y_ram_t = respuestas_tiempo(Hs)

# Pasamos las expresiones simb√≥licas a funciones de numpy para graficar
y_imp_fun = sym.lambdify(t, y_imp_t, "numpy")
y_esc_fun = sym.lambdify(t, y_esc_t, "numpy")
y_ram_fun = sym.lambdify(t, y_ram_t, "numpy")

# Vector de tiempo
t_vec = np.linspace(0, 10, 800)

tab_imp, tab_esc, tab_ram, tab_bode = st.tabs(
    ["Impulso", "Escal√≥n", "Rampa", "Bode / Polos"]
)

# ---- pesta√±a impulso ----
with tab_imp:
    st.write("Respuesta al impulso")
    fig1, ax1 = plt.subplots()
    ax1.plot(t_vec, y_imp_fun(t_vec))
    ax1.set_xlabel("t [s]")
    ax1.set_ylabel("y_impulso(t)")
    ax1.grid(True)
    st.pyplot(fig1)

# ---- pesta√±a escal√≥n ----
with tab_esc:
    st.write("Respuesta al escal√≥n unitario")
    fig2, ax2 = plt.subplots()
    ax2.plot(t_vec, y_esc_fun(t_vec))
    ax2.set_xlabel("t [s]")
    ax2.set_ylabel("y_escal√≥n(t)")
    ax2.grid(True)
    st.pyplot(fig2)

# ---- pesta√±a rampa ----
with tab_ram:
    st.write("Respuesta a la rampa")
    fig3, ax3 = plt.subplots()
    ax3.plot(t_vec, y_ram_fun(t_vec))
    ax3.set_xlabel("t [s]")
    ax3.set_ylabel("y_rampa(t)")
    ax3.grid(True)
    st.pyplot(fig3)

# ---- pesta√±a bode y polos/ceros ----
with tab_bode:
    st.write("Diagrama de Bode y polos/ceros")

    # Datos de bode
    w_vec, mag, fase = datos_bode(Hs)

    fig4, (axm, axp) = plt.subplots(2, 1, figsize=(6, 6))
    axm.semilogx(w_vec, mag)
    axm.set_ylabel("Magnitud [dB]")
    axm.grid(True, which="both")

    axp.semilogx(w_vec, fase)
    axp.set_xlabel("œâ [rad/s]")
    axp.set_ylabel("Fase [¬∞]")
    axp.grid(True, which="both")
    fig4.tight_layout()
    st.pyplot(fig4)

    # Polos y ceros
    ceros, polos = polos_y_ceros(Hs)
    fig5, ax5 = plt.subplots()
    if len(ceros) > 0:
        ax5.scatter(np.real(ceros), np.imag(ceros), marker='o', label="Ceros")
    ax5.scatter(np.real(polos), np.imag(polos), marker='x', color='red', label="Polos")
    ax5.axhline(0, color='black')
    ax5.axvline(0, color='black')
    ax5.set_xlabel("Re")
    ax5.set_ylabel("Im")
    ax5.grid(True)
    ax5.legend()
    ax5.set_title("Polos y ceros en el plano s")
    st.pyplot(fig5)

# ------------------ Tiempos caracter√≠sticos ------------------
st.subheader("Tiempos caracter√≠sticos de segundo orden (estimados)")

cTr, cMp, cTp, cTs = st.columns(4)
cTr.metric("t_r (levantamiento)", f"{tr:.4g} s" if not np.isnan(tr) else "N/A")
cMp.metric("M_p (sobreimpulso)", f"{Mp*100:.2f} %" if Mp > 0 else "0 %")
cTp.metric("t_p (pico)", f"{tp:.4g} s" if not np.isnan(tp) else "N/A")
cTs.metric("t_s (establecimiento 2%)", f"{ts:.4g} s" if not np.isnan(ts) else "N/A")
```
